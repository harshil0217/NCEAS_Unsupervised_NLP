# Load arXiv Dataset

```python
import pandas as pd
df = pd.read_csv("src/data/arxiv/arxiv_raw_30k.csv")
df_new = pd.DataFrame()
df_new["topic"] = df["text"]
df_new["category_1"] = df["label"].apply(lambda x: x.split(".")[0])
df_new["category_0"] = df["label"].apply(lambda x: x.split(".")[1])

df_new = df_new.dropna().reset_index(drop=True)

df_new
```

```python
df_new = df_new.dropna().reset_index(drop=True)

df_new = df_new[
    df_new["topic"].apply(lambda x: isinstance(x, str) and x.strip() != "")
].reset_index(drop=True)
```

```python
df_new.to_csv("src/data/arxiv/arxiv_clean.csv", index=False)
```
