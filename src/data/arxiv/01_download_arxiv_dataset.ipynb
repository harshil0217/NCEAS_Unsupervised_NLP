# arXiv Subject Taxonomy (CS / Physics) Download locally

```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("Cornell-University/arxiv")

print("Path to dataset files:", path)
```

```python
import os

os.listdir(path)
```

```python
import json
import os

# Construct the full path to the arXiv metadata snapshot file
# This file contains one JSON record per line (JSONL format),
# where each line corresponds to a single arXiv paper.
file_path = os.path.join(path, "arxiv-metadata-oai-snapshot.json")

# Open the file and read only the first line
# We intentionally load a single record to:
# 1) verify the dataset structure
# 2) confirm available fields (e.g., text + category labels)
# without loading the full multi-GB dataset into memory
with open(file_path, "r") as f:
    record = json.loads(f.readline())
    
    # Print the keys of a single record to inspect
    # what metadata fields are available in the dataset
    print(record.keys())
```

### Dataset Download (Local)

The Cornell arXiv dataset was downloaded locally using KaggleHub.  
Downloading the dataset locally means the files are stored on the local machine
and can be accessed directly for inspection and future processing without
re-downloading.

At this stage, the dataset is used only for validation purposes. A single record
is inspected to confirm the presence of textual fields (e.g., title, abstract)
and multi-label category metadata, without loading the full dataset into memory.


### Official arXiv Category Taxonomy

The hierarchical structure of arXiv subject categories is defined using the
official arXiv category taxonomy
(https://arxiv.org/category_taxonomy).

This taxonomy specifies parent–child relationships between top-level domains
(e.g., `cs`, `physics`) and their subcategories (e.g., `cs.AI`, `cs.LG`).
In this project, the taxonomy is used as a reference for defining hierarchical
relationships among category labels, rather than as a standalone dataset.
