```python
import os
import pandas as pd
from sklearn.datasets import fetch_rcv1
```

```python
os.makedirs("./data/", exist_ok=True)

def main():
    print("Fetching RCV1 from sklearn...")
    # This downloads the metadata/labels (~600MB)
    rcv1 = fetch_rcv1(subset='train', download_if_missing=True)
    
    target_names = rcv1.target_names
    labels = rcv1.target.toarray()
    
    data_list = []
    print("Building initial metadata table...")
    # Sampling 20,000 to keep it manageable
    for i in range(min(20000, len(rcv1.sample_id))):
        item_id = rcv1.sample_id[i]
        doc_labels = target_names[labels[i] == 1].tolist()
        
        data_list.append({
            "item_id": item_id,
            "topic": f"Placeholder text for article {item_id}", 
            "all_labels": "|".join(doc_labels)
        })

    df = pd.DataFrame(data_list)
    df.to_csv("./data/rcv1_processed.csv", index=False)
    print("Done! 'rcv1_processed.csv' is now ready for the Hierarchy script.")

if __name__ == "__main__":
    main()
```
