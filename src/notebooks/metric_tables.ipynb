```python
import pandas as pd
```

```python


def load_and_aggregate(path: str) -> pd.DataFrame:
    """
    Load a clustering scores CSV, drop 'level', aggregate by
    (embedding_model, reduction_method, cluster_method), and sort by AMI descending.
    """
    df = pd.read_csv(path)
    df = df.drop(columns=["level"])
    df = df.groupby(["embedding_model", "reduction_method", "cluster_method"]).mean().reset_index()
    df = df.sort_values(by="Dendrogram_Purity", ascending=False)
    return df
```
