```python
import os
import sys
from dotenv import load_dotenv
load_dotenv() 

# Set the target folder name you want to reach
target_folder = "phate-for-text"

# Get the current working directory
current_dir = os.getcwd()

# Loop to move up the directory tree until we reach the target folder
while os.path.basename(current_dir) != target_folder:
    parent_dir = os.path.abspath(os.path.join(current_dir, ".."))
    if parent_dir == current_dir:
        # If we reach the root directory and haven't found the target, exit
        raise FileNotFoundError(f"{target_folder} not found in the directory tree.")
    current_dir = parent_dir

# Change the working directory to the folder where "phate-for-text" is found
os.chdir(current_dir)

# Add the "phate-for-text" directory to sys.path
sys.path.insert(0, current_dir)
```

```python
import numpy as np
import phate
import numpy as np
from sklearn.decomposition import PCA
import umap
import os
```

```python
embedding_model = "MiniLM-L6-v2"
```

```python
data = np.load('data/epa/epa_embed.npy')

labels = np.load('data/epa/epa_labels.npy',allow_pickle=True)
```

```python
os.makedirs(f'{embedding_model}_reduced_embeddings', exist_ok=True)
```

```python
reducer_model = phate.PHATE(n_jobs=-2,random_state=42, n_components=3,decay=20,t="auto",n_pca=None) #{'k':10,'alpha':4,'t':3}
embed_phate = reducer_model.fit_transform(data)
np.save(f"{embedding_model}_reduced_embeddings/PHATE_epa_embed_top20.npy",embed_phate)
```

```python
embedding_methods = {}
embedding_methods["PHATE"]  =embed_phate

pca = PCA(n_components=3)
embedding_methods["PCA"] = pca.fit_transform(data)
np.save(f"{embedding_model}_reduced_embeddings/PCA_epa_embed_top20.npy",embedding_methods["PCA"])

umap_model = umap.UMAP(n_components=3
                       , random_state=42,min_dist=.05,n_neighbors=10)
embedding_methods["UMAP"] = umap_model.fit_transform(data)
np.save(f"{embedding_model}_reduced_embeddings/UMAP_epa_embed_top20.npy",embedding_methods["UMAP"])

from sklearn.manifold import TSNE

# # # Fit t-SNE
tsne_model = TSNE(n_components=3, random_state=42)
embedding_methods["T-SNE"] = tsne_model.fit_transform(data)
np.save(f"{embedding_model}_reduced_embeddings/tSNE_epa_embedtop20.npy",embedding_methods["T-SNE"])
```
