# Clustering Metric Summary Tables

Aggregated benchmark and synthetic clustering scores by reduction method, embedding model, and cluster method. Loads precomputed results CSVs from `results/`.

```python
import pandas as pd
```

```python


def load_and_aggregate(path: str) -> pd.DataFrame:
    """
    Load a clustering scores CSV, drop 'level', aggregate by
    (embedding_model, reduction_method, cluster_method), and sort by AMI descending.
    """
    df = pd.read_csv(path)
    df = df.drop(columns=["level"])
    df = df.groupby(["embedding_model", "reduction_method", "cluster_method"]).mean().reset_index()
    df = df.sort_values(by="Dendrogram_Purity", ascending=False)
    return df
```

```python
#combine all dfs into a single dataframe
benchmark_combined = pd.concat([amz, db, arxiv, rcv1, wos])
```

```python
#drop cluster method to isolate reduction method x embedding model
benchmark_combined = benchmark_combined.drop(columns = ['cluster_method', 'TED', 'embedding_model'], errors='ignore')
```

```python
#Group by reduction method + embedding model and take mean, then pivot embedding_model to top-level columns
benchmark_combined = benchmark_combined.groupby(['reduction_method']).mean()
benchmark_combined = benchmark_combined.sort_index(axis=1)
```

```python
benchmark_combined
```

### Average Clustering Method Performance (averaged across all benchmark datasets)

```python
benchmark_by_cluster = (
    pd.concat([amz, db, arxiv, rcv1, wos])
    .drop(columns=["TED", "embedding_model", "reduction_method"], errors="ignore")
    .groupby(["cluster_method"])
    .mean()
    .sort_index(axis=1)
)
benchmark_by_cluster
```

### Average Embedding Model Performance (averaged across all benchmark datasets)

```python
benchmark_by_embed = (
    pd.concat([amz, db, arxiv, rcv1, wos])
    .drop(columns=["TED", "cluster_method", "reduction_method"], errors="ignore")
    .groupby(["embedding_model"])
    .mean()
    .sort_index(axis=1)
)
benchmark_by_embed
```

### Top 8 Benchmark Combinations (averaged across all benchmark datasets)

```python
benchmark_top8 = (
    pd.concat([amz, db, arxiv, rcv1, wos])
    .drop(columns=["TED"], errors="ignore")
    .groupby(["embedding_model", "reduction_method", "cluster_method"])
    .mean()
    .reset_index()
    .sort_values("Dendrogram_Purity", ascending=False)
    .head(8)
    .reset_index(drop=True)
)
benchmark_top8
```
