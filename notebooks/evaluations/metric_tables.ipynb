# Clustering Metric Summary Tables

Aggregated benchmark and synthetic clustering scores by reduction method, embedding model, and cluster method. Loads precomputed results CSVs from `results/`.

```python
import pandas as pd
```

```python


def load_and_aggregate(path: str) -> pd.DataFrame:
    """
    Load a clustering scores CSV, drop 'level', aggregate by
    (embedding_model, reduction_method, cluster_method), and sort by AMI descending.
    """
    df = pd.read_csv(path)
    df = df.drop(columns=["level"])
    df = df.groupby(["embedding_model", "reduction_method", "cluster_method"]).mean().reset_index()
    df = df.sort_values(by="Dendrogram_Purity", ascending=False)
    return df
```

### Hercules - Benchmark Datasets

```python
amz_herc = load_and_aggregate("../../results/clustering/benchmark/amazon_herc_clustering_scores_direct.csv")
arxiv_herc = load_and_aggregate("../../results/clustering/benchmark/arxiv_herc_clustering_scores_direct.csv")
db_herc = load_and_aggregate("../../results/clustering/benchmark/db_herc_clustering_scores_direct.csv")
rcv1_herc = load_and_aggregate("../../results/clustering/benchmark/rcv1_herc_clustering_scores_direct.csv")
wos_herc = load_and_aggregate("../../results/clustering/benchmark/wos_herc_clustering_scores_direct.csv")
```

```python
# Average Hercules performance across all benchmark datasets, grouped by embedding model
benchmark_herc_combined = (
    pd.concat([amz_herc, arxiv_herc, db_herc, rcv1_herc, wos_herc])
    .drop(columns=["TED", "reduction_method", "cluster_method"], errors="ignore")
    .groupby(["embedding_model"])
    .mean()
    .sort_index(axis=1)
)
benchmark_herc_combined
```

```python
# Per-dataset Hercules scores (averaged across hierarchy levels) for each embedding model
herc_per_dataset = pd.DataFrame()
for name, df in [("amazon", amz_herc), ("arxiv", arxiv_herc), ("dbpedia", db_herc), ("rcv1", rcv1_herc), ("wos", wos_herc)]:
    tmp = df.drop(columns=["TED", "reduction_method", "cluster_method"], errors="ignore").copy()
    tmp.insert(0, "dataset", name)
    herc_per_dataset = pd.concat([herc_per_dataset, tmp])
herc_per_dataset = herc_per_dataset.sort_values(["dataset", "embedding_model"]).reset_index(drop=True)
herc_per_dataset
```

### Hercules vs. Standard Pipeline Comparison

```python
# Compare best standard pipeline vs Hercules for benchmark datasets
std_best = (
    pd.concat([amz, db, arxiv, rcv1, wos])
    .drop(columns=["TED", "embedding_model", "cluster_method", "reduction_method"], errors="ignore")
    .mean()
    .rename("Standard Pipeline (mean all)")
)
herc_mean = (
    pd.concat([amz_herc, arxiv_herc, db_herc, rcv1_herc, wos_herc])
    .drop(columns=["TED", "embedding_model", "cluster_method", "reduction_method"], errors="ignore")
    .mean()
    .rename("Hercules Direct (mean all)")
)
pd.DataFrame([std_best, herc_mean]).sort_index(axis=1)
```

```python
# Compare best standard pipeline vs Hercules for synthetic datasets
synth_std_best = (
    pd.concat([ecosystems_deep, ecosystems_shallow, fisheries_deep, fisheries_shallow])
    .drop(columns=["TED", "DP_Lower", "DP_Upper", "LCA_F1_Lower", "LCA_F1_Upper",
                    "embedding_model", "cluster_method", "reduction_method"], errors="ignore")
    .mean()
    .rename("Standard Pipeline (mean all)")
)
synth_herc_mean = (
    pd.concat([eeh_deep_herc, eeh_shallow_herc, eeh_shallow_noise_herc, oef_shallow_herc])
    .drop(columns=["TED", "DP_Lower", "DP_Upper", "LCA_F1_Lower", "LCA_F1_Upper",
                    "embedding_model", "cluster_method", "reduction_method"], errors="ignore")
    .mean()
    .rename("Hercules Direct (mean all)")
)
pd.DataFrame([synth_std_best, synth_herc_mean]).sort_index(axis=1)
```
